```python
import numpy as np
a=list()
b=list()
for i in range(0,5):
    a.append(int(input()))
for i in range(0,5):
    b.append(int(input()))
# convert them to array
A1=np.array(a)
A2=np.array(b)
print(A1)
print(A2)
# concatenate and print array
print(np.concatenate((A1,A2),axis=None))
# Sorting arrays
print(np.sort(A1))
print(np.sort(A2))
```
